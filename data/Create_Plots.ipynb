```python
# Per-label performance (precision, recall, F1) for CHAR, LOC, and ORG labels.

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
file_name = 'EWT-training-report.csv'
data = pd.read_csv(file_name)

# Filter the data to include only CHAR, LOC, and ORG labels
filtered_data = data[data['label'].isin(['CHAR', 'LOC', 'ORG'])]

# Set the labels for the x-axis
labels = filtered_data['label']

# Set the values for precision, recall, and F1-score
precision = filtered_data['precision']
recall = filtered_data['recall']
f1_score = filtered_data['f1-score']

# Define the bar width
bar_width = 0.25

# Set the position of the bars on the x-axis
r1 = range(len(labels))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plot
plt.figure(figsize=(10, 6))
bars1 = plt.bar(r1, precision, color='b', width=bar_width, edgecolor='grey', label='Precision')
bars2 = plt.bar(r2, recall, color='g', width=bar_width, edgecolor='grey', label='Recall')
bars3 = plt.bar(r3, f1_score, color='r', width=bar_width, edgecolor='grey', label='F1-score')

# Add labels and title
plt.xlabel('Labels', fontweight='bold')
plt.ylabel('Scores', fontweight='bold')
plt.title('Per-label Performance for CHAR, LOC, and ORG Labels', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(labels))], labels)

# Add legend
plt.legend()

# Add the values on top of the bars
def add_values(bars):
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.02, round(yval, 2), ha='center', va='bottom', fontweight='bold')

add_values(bars1)
add_values(bars2)
add_values(bars3)

# Show the plot
plt.show()

# Uncomment the line below to save the figure
plt.savefig(os.path.join(output_dir, 'per_label_performance.png'))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Baseline model', 'LOTR-trained model', 'EWT + LOTR-Trained Model']
metrics = ['Accuracy', 'F1 Score', 'CHAR', 'LOC', 'ORG']
baseline = [0.956, 0.513, 0.476, 0.652, 0.0]
lotr_trained = [0.991, 0.8814, 0.860, 0.960, 1.0]
ewt_lotr_trained = [0.990, 0.8757, 0.8549, 0.9722, 0.666]

# Bar chart parameters
bar_width = 0.2
x = np.arange(len(metrics))

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Creating bars for each model
bars1 = ax.bar(x - bar_width, baseline, width=bar_width, label='Baseline model')
bars2 = ax.bar(x, lotr_trained, width=bar_width, label='LOTR-trained model')
bars3 = ax.bar(x + bar_width, ewt_lotr_trained, width=bar_width, label='EWT + LOTR-Trained Model')

# Adding labels and title
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Adding data labels on top of the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# Display the plot
plt.tight_layout()
plt.show()
```
