```python
import pandas as pd
import matplotlib.pyplot as plt
import os
# Load the CSV file
file_path = 'EWT_epochs_metrics.csv'  # Update this path if needed
output_dir = 'plots'
df = pd.read_csv(file_path)
# Plot the F1 score over epochs
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['f1_score'], marker='o', linestyle='-', color='b')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.title('F1 Score Over Epochs')
plt.xticks(df['epoch'])
plt.grid(True)
plt.show()
plt.savefig('plots/EWT_epochs_metrics.jpg')

output_path = os.path.join(output_dir, 'EWT-epoch-trainings.png')
plt.savefig(output_path)
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the CSV file
file_path = 'metrics.csv'  # Update this path if needed
df = pd.read_csv(file_path)
df = df.sort_values(by='subset_size')

# Create the plots directory if it doesn't exist
output_dir = 'plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Plot the F1 score over subset sizes
plt.figure(figsize=(10, 6))
plt.plot(df['subset_size'], df['f1_score'], marker='o', linestyle='-', color='b')
plt.xlabel('Subset Size', fontsize=18)
plt.ylabel('F1 Score', fontsize=18)
plt.title('F1 Score over Subset Size', fontsize=20)
plt.grid(True)
# plt.show()
# Save the plot
plt.savefig(os.path.join(output_dir, 'F1_subset.png'))
plt.close()
```
